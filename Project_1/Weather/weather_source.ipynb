```python
#dependencies
import requests
import json
import datetime
import numpy as np
import pandas as pd
import os
from config import w_api_key
```

```python
#create empty dataframe
column_names = ['postal_code',
 'country',
 'timestamp',
 'cldCvrMin',
 'cldCvrAvg',
 'cldCvrMax',
 'dewPtMin',
 'dewPtAvg',
 'dewPtMax',
 'feelsLikeMin',
 'feelsLikeAvg',
 'feelsLikeMax',
 'heatIndexMin',
 'heatIndexAvg',
 'heatIndexMax',
 'mslPresMin',
 'mslPresAvg',
 'mslPresMax',
 'precip',
 'presTendMin',
 'presTendAvg',
 'presTendMax',
 'radSolarMin',
 'radSolarAvg',
 'radSolarMax',
 'radSolarTot',
 'relHumMin',
 'relHumAvg',
 'relHumMax',
 'sfcPresMin',
 'sfcPresAvg',
 'sfcPresMax',
 'snowDepth',
 'snowfall',
 'spcHumMin',
 'spcHumAvg',
 'spcHumMax',
 'tempMin',
 'tempAvg',
 'tempMax',
 'windChillMin',
 'windChillAvg',
 'windChillMax',
 'windDirAvg',
 'windDir80mAvg',
 'windDir100mAvg',
 'windSpdMin',
 'windSpdAvg',
 'windSpdMax',
 'windSpd80mMin',
 'windSpd80mAvg',
 'windSpd80mMax',
 'windSpd100mMin',
 'windSpd100mAvg',
 'windSpd100mMax',
 'wetBulbMin',
 'wetBulbAvg',
 'wetBulbMax']
```

```python
#create variables for the api url
country = 'US'
postal_code = '10019'
#define the API pull function
def pull(start,end):
    base_df = pd.DataFrame(columns = column_names)
    #set the start and end dates for months in the form of a list
    range_months = pd.date_range(start,end,freq='1M')
    #for each day from x/x/20xx-x/x/20xx ...
    for day in range_months:
        day_start = (day).strftime('%Y-%m-%d')
        day_end = (day+1*day.freq).strftime('%Y-%m-%d')
        try:
        #make the api call
            r = requests.get(f'https://api.weathersource.com/v1/{w_api_key}/postal_codes/{postal_code},{country}/history.json?period=day&timestamp_between={day_start}T00:00,{day_end}T00:00')
            read = r.json()
            #check time interval, and length to verify that we are getting the correct amount of information.
            print(f'timestamp_between={day_start}T00:00,{day_end}T00:00,length is:{len(read)}')
            #create dummy to temporarily store the dataframe
            dummy_df = pd.DataFrame(read,index=np.arange(start=0,stop=len(read)))
            base_df = base_df.append(dummy_df,ignore_index=False)
        except KeyError:
            print('Error, nothing found, moving to next')
    return base_df
```

```python
#calling def with 10 month intervals, waiting 5 min in between calls
#check length of each pull
w_1_15_10_15 = pull('01-01-2015','09-30-2015')
```

```python
#2nd 10-month interval
w_11_15_8_16 = pull('11-01-2015','08-30-2016')
```

```python
#3rd 10-month interval
w_9_16_6_17 = pull('09-01-2016','06-30-2017')
```

```python
#4th 10-month interval
w_8_17_7_18 = pull('08-01-2017','05-30-2018')
```

```python
#5th 10-month interval
w_6_18_3_19 = pull('06-01-2018','03-30-2019')
```

```python
#6th 10-month interval
w_4_19_1_20 = pull('04-01-2019','01-30-2020')
```

```python
#final 10-month interval
w_2_20_7_20 = pull('02-01-2020','07-25-2020')
```

```python
#combine dataframes
weather_df = pd.DataFrame(columns = column_names)
weather_df = weather_df.append([w_2_20_7_20, w_4_19_1_20,w_6_18_3_19,w_8_17_7_18,w_9_16_6_17,w_11_15_8_16,w_1_15_10_15],ignore_index=True)
weather_df
```

```python
#truncate data and rename labels
weather_truncate = weather_df[['timestamp','tempAvg','relHumAvg','snowDepth','windSpdAvg','precip','tempMax','windSpdMax']]
weather = weather_truncate.rename(columns={"timestamp": "Day",
                            "tempAvg": "Temperature (F)",
                            "relHumAvg": "Humitdity (%)",
                            "snowDepth": "Snowfall Depth (in)",
                            "windSpdAvg": "Wind Speed (MPH)",
                            "precip": "Precipitation (in)",
                            'tempMax':'Max Temperature (F)',
                            'windSpdMax':'Max Wind Speed (MPH)'})
weather
```

```python
#change days to month by slicing off datetime stopping on month
weather_month = weather
weather_month['Day'] = weather_month['Day'].str.slice(start=0,stop=7)
weather_month = weather_month.rename(columns={'Day':'Month'})
weather_month
```

```python
#export data to csv 
weather_month.to_csv('weather_by_month.csv')
```
