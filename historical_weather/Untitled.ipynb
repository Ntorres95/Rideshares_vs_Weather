```python
import pandas as pd
import os
import json
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# data = 'FHV_Base_Aggregate_Report.csv'
# trip_data_df = pd.read_csv(data)
# trip_data_df.head()
city_attributes = 'city_attributes.csv'
humidity = 'humidity.csv'
pressure = 'pressure.csv'
temp = 'temperature.csv'
weather_des = 'weather_description.csv'
wind_speed = 'wind_speed.csv'

city_attributes_df = pd.read_csv(city_attributes)
humidity_df = pd.read_csv(humidity)
pressure_df = pd.read_csv(pressure)
temp_df = pd.read_csv(temp)
weather_des_df = pd.read_csv(weather_des)
wind_speed_df = pd.read_csv(wind_speed)
```

```python
city_attributes_df.head()
# lyft_df = clean_df.loc[clean_df["Base Name"] == "LYFT", :]
ny_city = city_attributes_df.loc[city_attributes_df["City"]== "New York", :]
ny_city
```

```python
ny_humidity_df = humidity_df[["datetime", "New York"]]
ny_humidity_df.head()
ny_humidity_df = ny_humidity_df.dropna(how="any")
ny_humidity_df.head()
```

```python
pressure_df.head()
```

```python
ny_pressure_df = pressure_df[["datetime", "New York"]]
ny_pressure_df.head()
ny_pressure_df = ny_pressure_df.dropna(how="any")
ny_pressure_df.head()
```

```python
temp_df.head()
```

```python
ny_temp_df = temp_df[["datetime", "New York"]]
ny_temp_df.head()
ny_temp_df = ny_temp_df.dropna(how="any")
ny_temp_df.head()
```

```python
weather_des_df.head()
```

```python
ny_weather_df = weather_des_df[["datetime", "New York"]]
ny_weather_df.head()
ny_weather_df = ny_weather_df.dropna(how="any")
ny_weather_df.head()
```

```python
wind_speed_df.head()
```

```python
ny_wind_speed_df = wind_speed_df[["datetime", "New York"]]
ny_wind_speed_df.head()
ny_wind_df = ny_wind_speed_df.dropna(how="any")
ny_wind_df.head()
```

```python

new_york_df = new_york_df.rename(columns={"New York_x": "NY Humidity",
                                         "New York_y": "NY Pressure"})
new_york_df.head()
```

```python
new_york_df = pd.merge(new_york_df, ny_temp_df, how='outer', on="datetime")
new_york_df
```
